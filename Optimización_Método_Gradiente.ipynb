```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_a_optimizar(x):
    return x**2 - 4*x + 4 #Función fija

def derivada_funcion(x):
    return 2*x - 4  # Derivada de la función anterior.
```

```python
def descenso_de_gradiente(inicio_x, tasa_aprendizaje, iteraciones):
    puntos_x = [inicio_x]
    puntos_y = [funcion_a_optimizar(inicio_x)]

    for i in range(iteraciones):
        gradiente = derivada_funcion(inicio_x)
        inicio_x = inicio_x - tasa_aprendizaje * gradiente  # Actualiza x utilizando la tasa de aprendizaje y el gradiente
        puntos_x.append(inicio_x)
        puntos_y.append(funcion_a_optimizar(inicio_x))

    return puntos_x, puntos_y
```

```python

def visualizar_optimizacion(puntos_x, puntos_y):
    x = np.linspace(min(puntos_x) - 1, max(puntos_x) + 1, 400)
    y = funcion_a_optimizar(x)

    plt.plot(x, y, '-g', label='Función a optimizar')
    plt.scatter(puntos_x, puntos_y, c='b', marker='o', label='Pasos del descenso de gradiente')
    plt.title('Optimización utilizando Descenso de Gradiente')
    plt.xlabel('X')
    plt.ylabel('f(X)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def main():
    inicio_x = float(input("Valor inicial de x: "))
    tasa_aprendizaje = float(input("Tasa de aprendizaje (ejemplo: 0.1): "))
    iteraciones = int(input("Número de iteraciones: "))

    puntos_x, puntos_y = descenso_de_gradiente(inicio_x, tasa_aprendizaje, iteraciones)
    visualizar_optimizacion(puntos_x, puntos_y)

if __name__ == "__main__":
    main()
```
